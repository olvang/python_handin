```python
class Student():

    def __init__(self, name, gender, datasheet, image_url):
        self.name = name 
        self.gender = gender
        self.datasheet = datasheet
        self.image_url = image_url
      
    def get_avg_grade(self):
        grades = self.datasheet.get_grades_as_list()
        if(len(grades) > 0):
            return sum(grades) / len(grades)
        else:
            return 'No grades to calculate avg'
        
    def get_progression(self):
        passedCourses = self.datasheet.get_passed_corses()
        totalETCS = 0
        for course in passedCourses:
            totalETCS += course.ETCS
        return (totalETCS / 150) * 100

class DataSheet():

    def __init__(self, courses=[]):
        self.courses = courses
    
    def get_grades_as_list(self):
        grades = []
        for course in self.courses:
            if(hasattr(course, 'grade') and course.grade != "None"):
                grades.append(course.grade)
        return grades
    
    def get_passed_corses(self):
        courses = []
        for course in self.courses:
            if(course.grade != "None" and course.grade > 0):
                courses.append(course)
        return courses
                
class Course():

    def __init__(self, name, classroom, teacher, ETCS, grade="None"):
        self.name = name
        self.classroom = classroom
        self.teacher = teacher
        self.ETCS = ETCS
        self.grade = grade
```

```python
# EX 1: 1-6
course1 = Course('Course1', 'A', 'Teacher1', 10, 4)
course2 = Course('Course2', 'B', 'Teacher2', 20, 7)
course3 = Course('Course3', 'C', 'Teacher3', 30, 10)
course4 = Course('Course4', 'D', 'Teacher4', 40, 12)
courses = [course1,course2,course3,course4]
datasheet1 = DataSheet(courses)

student1 = Student('Name1', 'Male', datasheet1, 'https://picsum.photos/200')
avg_grade = student1.get_avg_grade()
print(avg_grade)
```

```python
#EX 1: 7
import random
import csv

first_names=("Harry","Ross","Bruce","Cook","Carolyn","Morgan","Albert","Walker","Randy","Reed","Larry","Barnes","Lois","Wilson","Jesse","Campbell","Ernest","Rogers","Theresa","Patterson","Henry","Simmons","Michelle","Perry","Frank","Butler","Shirley")
last_names=("Brooks","Rachel","Edwards","Christopher","Perez","Thomas","Baker","Sara","Moore","Chris","Bailey","Roger","Johnson","Marilyn","Thompson","Anthony","Evans","Julie","Hall","Paula","Phillips","Annie","Hernandez","Dorothy","Murphy","Alice","Howard")
grades = ['None',-3, 0, 2, 4,7,10,12]
genders = ['MALE', 'FEMALE']
profilePictures = ['https://picsum.photos/id/1/200/200','https://picsum.photos/id/2/200/200','https://picsum.photos/id/3/200/200','https://picsum.photos/id/4/200/200','https://picsum.photos/id/5/200/200']

def generateCourses(courseNames):
    courses = []
    nCourses = random.randrange(1,len(courseNames))
    ranNames = random.choices(courseNames, k=nCourses)
    for i in range(nCourses):
        classroom = chr(random.randrange(97, 97 + 26))
        teacher = random.choice(first_names) + ' ' + random.choice(last_names)
        ETCS = random.randrange(40)
        grade = random.choice(grades)
        course = Course(ranNames[i], classroom, teacher, ETCS, grade)
        courses.append(course)
    return courses

def generateStudents(courseNames, nStudents=5):
    students = []
    for i in range(nStudents):
        courses = generateCourses(courseNames)
        datasheet = DataSheet(courses)
        
        name = random.choice(first_names) + ' ' + random.choice(last_names)
        student = Student(name, random.choice(genders), datasheet, random.choice(profilePictures))
        students.append(student)
    return students
    
    
courseNames = ["Hindus, Cosmology, and Petrology", "Design and Advocacy in Mental Health and Illness", "Purpose, Culture, and Diversity in Kiswahili Literature", "Morphology, Citizens, and Caste", "Ethnicity and Optics in Conflict and Transition", "Media, Performance, and Foxes in the Sixteenth Century", "Anatomy, Application, and Thermodynamics", "Decay, Sustainability, and Change in Gothic Europe", "Creativity, Fields, and Evolutionary"]
students = generateStudents(courseNames, 5)

# 7a
with open('students.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    for student in students:
        for course in student.datasheet.courses:
            writer.writerow([student.name, course.name, course.teacher, course.ETCS, course.classroom, course.grade, student.image_url])
```

```python
import csv
%matplotlib notebook
import matplotlib.pyplot as plt

# ex1: 8
with open('students.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    students = []
    for row in csv_reader:
        grade = 'None'
        try:
            grade = int(row[5])
        except:
            pass
        course = Course(row[1], row[4], row[2], int(row[3]), grade)
        if(len(students) == 0 or row[0] != students[-1].name):
            #new student
            datasheet = DataSheet([course])
            student = Student(row[0], "UNKNOWN",datasheet, row[6])
            students.append(student)
        else:
            #student same as last, add course to student
            courses = students[-1].datasheet.courses
            
            courses.append(course)
            students[-1].datasheet.courses = courses
    
    # A
    for student in students:
        avgGrade = student.get_avg_grade();
        print('Name: {} image: {} avgGrade: {}'.format(student.name, student.image_url, avgGrade))
    
    #B
    print('SORTED ----')
    studentsSorted = sorted(students, key=lambda student: student.get_avg_grade() ,reverse=True)
    for student in studentsSorted:
        avgGrade = student.get_avg_grade();
        print('Name: {} image: {} avgGrade: {} passed: {}'.format(student.name, student.image_url, avgGrade, student.get_progression()))
    
    #C
    print('CHART ----')
    names = []
    avg_grades = []
    for student in students:
        names.append(student.name)
        avg_grades.append(student.get_avg_grade())
    plt.bar(names,avg_grades,width=0.5, align='center')
    plt.xticks(rotation=45, horizontalalignment='right',fontweight='light')
```
